```python
# Define a double dimensional array (n x n), print it's transpose matrix and print another 2D matrix (2 X n) which prints the sum of
#rows and columns.
import numpy as np

n = int(input("Enter the size of the square matrix (n x n): "))
print("Enter the elements of the matrix row by row, separated by spaces:")
matrix = np.array([list(map(float, input().split())) for _ in range(n)])

print("\nOriginal Matrix:")
print(matrix)

transpose_matrix = matrix.T
print("\nTranspose Matrix:")
print(transpose_matrix)

row_sums = matrix.sum(axis=1)
col_sums = matrix.sum(axis=0)
sum_matrix = np.vstack([row_sums, col_sums])

print("\n2D Matrix (2 x n) showing the sum of rows and columns:")
print(sum_matrix)
```

```python
#Plot graphs of relu, sigmoid, and tanh activation functions.(relu on different graph while the rest 2 on same graph from -100 to 100).
#Just go through their definitions as of now.
import matplotlib.pyplot as plt
x = np.linspace(-100, 100, 1000)

relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)

plt.figure(figsize=(6, 4))
plt.plot(x, relu, label='ReLU', color='blue')
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(x, sigmoid, label='Sigmoid', color='green')
plt.plot(x, tanh, label='Tanh', color='orange')
plt.title('Sigmoid and Tanh')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
